```python
#---------------------------------------------------------------------------------
#transform auxiliary values (which are bounded [0,1] into the breakpoint dates)
def aux_to_breakpoints(aux, x0, xn, minwindow):
    breakpoints = []
    bi1 = x0
    n=len(aux)
    for i,a in enumerate(aux):
        bi = a * (xn-(n-i)*minwindow - (bi1+minwindow)) + bi1+minwindow
        breakpoints.append(bi)
        bi1=bi   
    return breakpoints

#---------------------------------------------------------------------------------
def breakpoints_to_aux(breakpoints, x0, xn, minwindow):
    aux = []
    bi1 = x0
    n=len(breakpoints)
    for i,b in enumerate(breakpoints):
        temp = xn - (n-i)*minwindow - (bi1+minwindow)
        a = 1.0 if temp==0 else (b - bi1 - minwindow) / temp
        aux.append(a)
        bi1=b
    return aux  
    
#aux_to_breakpoints([0.5, 0.5], 0, 100, 10)


#---------------------------------------------------------------------------------


def override(key, overrides, default):
    return default if key not in overrides else overrides[key]
```

```python
def modelfunc_SISV2(x, params, constants):
    #params are the values of beta0, beta1, ...
    #print(constants)
    constants['beta0'] = params[0]
    for i in range(1, constants['segments']+1):
        constants['beta{}'.format(i)] = params[i]
        
    yhat, yyhat = s.SISV2(x, constants)
     
    return yhat[:,s.cF]
```

```python
def diffevol_inner(aux, constants):
    #aux are the values of t1, ...
    x = constants['x']
    y = constants['y']
    bounds = constants['bounds']
    guess = constants['guess']
    alpha = constants['alpha']
    
    params = constants['params']
    
    breakpoints = aux_to_breakpoints(aux, x[0], x[-1], constants['minwindow']) 
    for i, t in enumerate(breakpoints):
        params['t{}'.format(i+1)] = t

    params, likelihood, fit = fit_model(x, y, model_func=modelfunc_SISV2, constants=params, 
                                        loglik_func=loglik_leastsquare, 
                                        guess=guess, bounds=bounds, alpha=alpha)
    
    print(f"{likelihood:,}", aux,breakpoints)
    
    return likelihood
```

```python
def SISV2_diffevol(x, y, params):
    
    segments = params['segments']  #0 means constant contact rate
    gamma = params['gamma']
    
    #prepare the bounds and guess vectors, and the other constants that will be used in the inner optimizer
    guess = [2*gamma]   #beta0
    bounds = [(1.1*gamma, 10*gamma)]  #bounds for beta0
    alpha = 1
    
    for i in range(segments):
        guess.append(1.3*gamma)  #growth rate after breakpoint
        bounds.append((0.1*gamma,5*gamma))
    
    constants = {'minwindow': params['minwindow'], 'x':x, 'y':y, 'guess':guess, 'bounds':bounds, 'alpha':alpha, 'params':params}
    
    #bounds for the auxiliary variables that will be used to set the breakpoints
    #these auxiliary variables are optimized by differential_evolution()
    #auxBounds = [(0,1)] * (segments+1)
    auxBounds = [(0,0.1)]
    for i in range(segments):
        auxBounds.append((0,1))
    
    #find the optimal breakpoints
    res = differential_evolution(func=diffevol_inner, bounds=auxBounds, args=[constants], updating='immediate')
    #print(res)
    
    #do the regression for the optimum breakpoints
    aux = res.x
    breakpoints = aux_to_breakpoints(aux, x[0], x[-1], params['minwindow']) 
    for i, t in enumerate(breakpoints):
        params['t{}'.format(i+1)] = t
        
    p, likelihood, fit = fit_model(x, y, model_func=modelfunc_SISV2, constants=params, 
                                        loglik_func=loglik_leastsquare, 
                                        guess=guess, bounds=bounds, alpha=alpha)

    params['beta0'] = p[0]
    for i in range(1,segments+1):
        params['beta{}'.format(i)] = p[i]
    
    return params, likelihood, fit


```

```python
def fit_giventimes(x, y, params):
    
    segments = params['segments']  #0 means constant contact rate
    gamma = params['gamma']
    
    #prepare the bounds and guess vectors, and the other constants that will be used in the inner optimizer
    guess = [2*gamma]   #beta0
    bounds = [(1.1*gamma, 10*gamma)]  #bounds for beta0
    alpha = 1
    
    for i in range(segments):
        guess.append(1.3*gamma)  #growth rate after breakpoint
        bounds.append((0.1*gamma,5*gamma))
   
    constants = {'minwindow': params['minwindow'], 'x':x, 'y':y, 'guess':guess, 'bounds':bounds, 'alpha':alpha, 'params':params}
 
    p, likelihood, fit = fit_model(x, y, model_func=modelfunc_SISV2, constants=params, 
                                    loglik_func=loglik_leastsquare, 
                                    guess=guess, bounds=bounds, alpha=alpha)

    params['beta0'] = p[0]
    for i in range(1,segments+1):
        params['beta{}'.format(i)] = p[i]
        
    return params, likelihood, fit
```
